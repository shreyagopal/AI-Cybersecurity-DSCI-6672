# **Perceptron for Phish Detection**

**Task:** Train a single-layer perceptron model to create a phish detector. You will use the Phishing Website dataset ( https://archive.ics.uci.edu/ml/datasets/phishing+websites ) for training and evaluating your model. The description of features can be found in the Data Folder. 

```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating dataframe for the phishing dataset
df = pd.read_csv('phishing_dataset.csv',header = None)
df
```

```
# Creating target & sample variables

y = df.iloc[:, 30].values
X = df.iloc[:, :30].values
y.shape, X.shape
```

```
# Splitting the data into train and test datasets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
         X, y, test_size=0.2, random_state= 12) 
X_test.shape, X_train.shape
```

```
# Fitting the train data to perceptron model
from sklearn.linear_model import Perceptron

p = Perceptron(max_iter=5000, eta0=0.0003, random_state=4)
p.fit(X_train, y_train)
```

```
# Predicting the target value for the test dataset

y_pred = p.predict(X_test)
```

```
# Calculating accuracy and number of mispredicted samples

from sklearn.metrics import accuracy_score

print('Misclassified samples: %d' % (y_test != y_pred).sum())
print('Accuracy: %.5f' % accuracy_score(y_test, y_pred))
```
